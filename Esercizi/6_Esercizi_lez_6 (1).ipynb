### **1. Distribuzioni di probabilità**
Genera 1000 valori casuali da una distribuzione gamma con parametro di forma pari a 1.
Suggerimento: il parametro di forma viene passato come primo argomento quando si "congela" la distribuzione.

Traccia l’istogramma del campione e sovrapponi la PDF della distribuzione.

Stima il parametro di forma dal campione usando il metodo fit.

Extra:
Le distribuzioni hanno molti metodi utili. Esplorali usando il completamento automatico con il tasto TAB.

Traccia la funzione di distribuzione cumulativa (CDF).

Calcola la varianza.

```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
distribuzione = sp.stats.gamma(1)
sample = distribuzione.rvs(size=1000)

plt.hist(sample, bins=50, density=True, label="istogramma normalizzato")
x = np.linspace(0,10,1000)
plt.plot(x, distribuzione.pdf(x), label="PDF gamma con  n=1")
plt.legend()
plt.show()
```

```python
forma, loc, scala = sp.stats.gamma.fit(sample)
forma, loc, scala
```

```python
plt.plot(x, distribuzione.cdf(x), label="CDF")
plt.legend()
plt.show()
```

```python
var = np.var(sample)
var
```

### **2. Fitta i dati**
Prova a fittare i dati sottostante con le migliori curve, calcola il MAE e l'RMSE

```python
import numpy as np
temp_max = np.array([17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18])
temp_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
import matplotlib.pyplot as plt
months = np.arange(12)
plt.plot(months, temp_max, "ro")
plt.plot(months, temp_min, "bo")
plt.xlabel("Month")
plt.ylabel("Min and max temperature")
```

```python
#provo usando sia numpy che scipy
#NUMPY
x_interpolare = np.linspace(0,11,12) #perchp il vvettore mesi è generato np.arange(12) che va 0,1,..,11 
coefficienti_np_max = np.polyfit( months, temp_max, deg=3)
intercette_np_max = np.polyval(coefficienti_np_max, x_interpolare)
```

```python
#SCIPY interp1d
funzione_sp_max = sp.interpolate.interp1d(months, temp_max,  kind=3)
intercette_sp_max = funzione_sp_max(x_interpolare)
```

```python
#SCIPY SMOOTH
funzione_spline_max = sp.interpolate.UnivariateSpline(months, temp_max, s=2)
intercette_spline_max = funzione_spline_max(x_interpolare)
```

```python
#ora faccio la stessa cosa per i minimi
coefficienti_np_min = np.polyfit(months, temp_min, deg=3)
intercette_np_min = np.polyval(coefficienti_np_min, x_interpolare)
```

```python
funzione_sp_min = sp.interpolate.interp1d(months, temp_min, kind=3)
intercette_sp_min = funzione_sp_min(x_interpolare)
```

```python
funzione_spline_min = sp.interpolate.UnivariateSpline(months, temp_min, s=3)
intercette_spline_min = funzione_spline_min(x_interpolare)
```

```python
plt.plot(months, temp_max, "ro")
plt.plot(months, temp_min, "bo")

#plot sui massimi
plt.plot(x_interpolare, intercette_np_max, label='polyfit-polyval', color='purple')

plt.plot(x_interpolare, intercette_sp_max, label='interp1d', color = 'red')

plt.plot(x_interpolare, intercette_spline_max, label='UnivariateSpline', color = 'green')

#plot sui minimi
plt.plot(x_interpolare, intercette_np_min, color='purple')

plt.plot(x_interpolare, intercette_sp_min, color='red')

plt.plot(x_interpolare, intercette_spline_min, color='green')


plt.xlabel("Month")
plt.ylabel("Min and max temperature")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#MAE
np.mean( abs(temp_max-intercette_spline_max))

#RMSE
(np.mean( (temp_max-intercette_spline_max)**2) )**1/2
```

### **3. Modello di regressione lineare dei seguenti dati**

 Calcola un modello di regressione lineare delle colonne mpg e disp del famoso dataset mtcars.

Dove: 

mpg = Miles Per Gallon → miglia per gallone, cioè una misura del consumo di carburante. Più alto è il valore, più efficiente è l’auto.

disp = Displacement → cilindrata del motore, in pollici cubici (cubic inches). Rappresenta il volume totale dei cilindri del motore. Più è alto, maggiore è la potenza potenziale del motore (ma anche il consumo).

```python
import pandas as pd

# Load dataset from URL
df = pd.read_csv('https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv')

# Display the dataframe
df
```

```python
#tramite numpy
m, q = np.polyfit(df["disp"], df["mpg"], 1)
reg_lin_np = m*df["disp"] + q
```

```python
#tramite scipy
coef, inter, r_sq, p_val, err = sp.stats.linregress(df.disp, df.mpg)
print(f"r-squared: {r_sq}")
print(f"p-value: {p_val}")

reg_lin_sp = coef*df["disp"] + inter
```

```python
plt.scatter(df.disp, df.mpg)
plt.plot(df.disp, reg_lin_np, label = 'regressione numpy')
plt.plot(df.disp, reg_lin_sp, label = 'regressione scipy', linestyle = ':')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
mae = np.mean( abs(df.mpg - reg_lin_sp))
rmse = ( np.mean( abs(df.mpg - reg_lin_sp)**2 ) )**1/2

mae, rmse
```
